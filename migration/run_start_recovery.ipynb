```python
%matplotlib notebook
from databroker import Broker
from metadatastore.mds import MDS
from filestore.fs import FileStore
from hxntools.handlers.xspress3 import Xspress3HDF5Handler
from hxntools.handlers.timepix import TimepixHDF5Handler
import time

from pymongo import MongoClient
import datetime
import numpy as np
import os
import pytz
import matplotlib.pyplot as plt

_mds_config = {'host': 'xf03id-ca1',
               'port': 27017,
               'database': 'datastore-new',
               'timezone': 'US/Eastern'}
mds = MDS(_mds_config, auth=False)
_fs_config = {'host': 'xf03id-ca1',
              'port': 27017,
              'database': 'filestore-new'}
db = Broker(mds, FileStore(_fs_config))
db.fs.register_handler(Xspress3HDF5Handler.HANDLER_NAME,
                       Xspress3HDF5Handler)
db.fs.register_handler(TimepixHDF5Handler._handler_name,
                       TimepixHDF5Handler, overwrite=True)

conn = MongoClient(host=_fs_config['host'],
                   port=_fs_config['port'])
```

```python
#34058 to 34064.
id_start = 34058
id_end = 34065
d1 = db[id_start]
d2 = db[id_end]
t1 = d1.start.time
t2 = d2.start.time
```

```python
d1.start.time
```

```python
m1 = mds.find_descriptors(start_time=t1)
```

```python
res = conn['datastore-new'].event_descriptor.find({'time':{'$lte':t2,'$gte':t1 }})
```

```python
from pymongo.errors import DuplicateKeyError
import uuid
sc_id = 34059
for r in res:
    try:
        mds.insert_run_start(uid=r['run_start'], time=r['time'], scan_id=sc_id, recovery_script=True)
        print('RunStart created %', r['run_start'])
        rstop = mds.insert_run_stop(run_start=r['run_start'], time=time.time(), recovery_script=True,
                                    exit_status='success',
                                   uid=str(uuid.uuid4()))
        print('RunStop created %s', rstop)
        sc_id += 1
        print('scan_ind is %i', sc_id)
    except DuplicateKeyError:
        print('caught duplicate')
```
